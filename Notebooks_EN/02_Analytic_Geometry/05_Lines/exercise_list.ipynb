### Problem: Find the Equation of the Line Passing Through Points \( A(1, 2) \) and \( B(3, 4) \)

---

### Step-by-Step Solution:

#### Step 1: Find the Slope (\( m \)) of the Line
The slope of a line passing through two points \( A(x_1, y_1) \) and \( B(x_2, y_2) \) is given by:
$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Substitute the coordinates of \( A(1, 2) \) and \( B(3, 4) \):
$$
m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1
$$

---

#### Step 2: Use the Point-Slope Form to Find the Equation
The point-slope form of a line is:
$$
y - y_1 = m(x - x_1)
$$

Using point \( A(1, 2) \) and slope \( m = 1 \):
$$
y - 2 = 1(x - 1)
$$

Simplify:
$$
y - 2 = x - 1
$$
$$
y = x - 1 + 2
$$
$$
y = x + 1
$$

---

#### Step 3: Write the Final Equation
The equation of the line in slope-intercept form is:
$$
\boxed{y = x + 1}
$$

---

   This matches the \( y \)-coordinate of \( A \).

2. For \( B(3, 4) \):

   This matches the \( y \)-coordinate of \( B \).

### Step 2: What Does "Perpendicular" Mean?
Two lines are **perpendicular** if they meet at a right angle (90 degrees). For their slopes:
- If the slope of one line is \( m \), the slope of the perpendicular line is the **negative reciprocal** of \( m \), which is \( -\frac{1}{m} \).

---

### Step 3: Find the Slope of the Given Line
The given line is \( y = 2x + 3 \). This is in the form \( y = mx + b \), where:
- \( m \) is the slope.
- \( b \) is the y-intercept.

So, the slope of the given line is:
$$
m_1 = 2
$$

---

### Step 4: Find the Slope of the Perpendicular Line
The slope of the perpendicular line is the **negative reciprocal** of \( m_1 \). So:
$$
m_2 = -\frac{1}{m_1} = -\frac{1}{2}
$$

---

### Step 5: Use the Point-Slope Form to Find the Equation
We know:
- The slope of the new line is \( m_2 = -\frac{1}{2} \).
- The line passes through the point \( A(1, 2) \).

The **point-slope form** of a line is:
$$
y - y_1 = m(x - x_1)
$$
Where:
- \( (x_1, y_1) \) is a point on the line.
- \( m \) is the slope.

Substitute \( m = -\frac{1}{2} \) and \( (x_1, y_1) = (1, 2) \):
$$
y - 2 = -\frac{1}{2}(x - 1)
$$

---

### Step 6: Simplify the Equation
Letâ€™s simplify the equation step by step:
1. Distribute \( -\frac{1}{2} \):
   $$

   $$
   y = -\frac{1}{2}x + \frac{1}{2} + 2
   $$

3. Combine the constants:
   $$
   y = -\frac{1}{2}x + \frac{5}{2}
   $$

---

### Final Answer:

## **Finding the Equation of a Line Through Two Points**

We need to find the equation of the line passing through \( A(2, 0) \) and \( B(0, 3) \).

---

### **Steps for Beginners**

### 1. **Understand the general line equation**  
The equation of a line is:  
\[
y = mx + b
\]  
Where:  
- \( m \) is the slope (how steep the line is).  
- \( b \) is the \( y \)-intercept (where the line crosses the \( y \)-axis).  

---

### 2. **Find the slope \( m \)**  
The slope \( m \) is found using the two points \( A(2, 0) \) and \( B(0, 3) \):  
\[
m = \frac{\text{change in } y}{\text{change in } x} = \frac{y_2 - y_1}{x_2 - x_1}
\]  
Substitute the points \( A(2, 0) \) (\( x_1 = 2, y_1 = 0 \)) and \( B(0, 3) \) (\( x_2 = 0, y_2 = 3 \)):  
\[
m = \frac{3 - 0}{0 - 2} = \frac{3}{-2} = -1.5
\]  

So, the slope \( m = -1.5 \).

---

### 3. **Write the line equation using \( m \) and \( b \)**  
The line equation is:  
\[
y = mx + b
\]  
We already know \( m = -1.5 \), so:  
\[
y = -1.5x + b
\]

---

### 4. **Find \( b \) (the \( y \)-intercept)**  
To find \( b \), use one of the points. Letâ€™s use \( B(0, 3) \), where \( x = 0 \) and \( y = 3 \):  
\[
3 = -1.5(0) + b
\]  
\[
b = 3
\]

---

### 5. **Write the final equation**  
Now we know \( m = -1.5 \) and \( b = 3 \):  
\[
y = -1.5x + 3
\]

---

### **Final Answer**  
The equation of the line through \( A(2, 0) \) and \( B(0, 3) \) is:  
\[
y = -1.5x + 3

### Problem:
Find a vector perpendicular to the line \( x + y + 1 = 0 \).

---

### Step 1: Understand the Line Equation
The given line is:
$$
x + y + 1 = 0
$$

This is in the standard form of a line:
$$
Ax + By + C = 0
$$
where:
- \( A = 1 \)
- \( B = 1 \)
- \( C = 1 \)

---

### Step 2: Find the Normal Vector
The coefficients \( A \) and \( B \) in the standard form of a line give us a vector that is **perpendicular** (normal) to the line. This vector is:
$$
\vec{n} = (A, B) = (1, 1)
$$

---

### Step 3: Verify Perpendicularity
To confirm that \( \vec{n} = (1, 1) \) is perpendicular to the line, we can check the **dot product** with the direction vector of the line.

1. **Direction Vector of the Line**:
   Rewrite the line \( x + y + 1 = 0 \) in slope-intercept form:
   $$
   y = -x - 1
   $$
   The slope of the line is \( -1 \), so the direction vector \( \vec{d} \) of the line is:
   $$
   \vec{d} = (1, -1)
   $$

2. **Dot Product**:
   The dot product of \( \vec{n} \) and \( \vec{d} \) is:
   $$
   \vec{n} \cdot \vec{d} = (1)(1) + (1)(-1) = 1 - 1 = 0
   $$
   Since the dot product is \( 0 \), the vectors are perpendicular.

---

### Final Answer:
A vector perpendicular to the line \( x + y + 1 = 0 \) is:
$$
\vec{n} = (1, 1)
$$

---

### Extra Notes:
- Any vector that is a multiple of \( \vec{n} = (1, 1) \) is also perpendicular to the line. For example:
  - \( \vec{n} = (2, 2) \)
  - \( \vec{n} = (-1, -1) \)
