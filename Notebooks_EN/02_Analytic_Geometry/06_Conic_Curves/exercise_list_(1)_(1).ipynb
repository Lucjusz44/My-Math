### **Problem 1: Equation of a Circle**
**Given:**  
- Center at \( A(1, 2) \)  
- Radius \( r = 3 \)  

**Equation of the Circle:**  
The standard equation of a circle is:  
$$(x - h)^2 + (y - k)^2 = r^2$$  
Substitute \( h = 1 \), \( k = 2 \), and \( r = 3 \):  
$$(x - 1)^2 + (y - 2)^2 = 9$$

### **Problem 2: Equation of a Parabola**
**Given:**  
- Intersects the \( x \)-axis at \( x = 2 \) and \( x = 4 \)  
- Passes through the point \( y(3) = 1 \)  

**Equation of the Parabola:**  
The parabola intersects the \( x \)-axis at \( x = 2 \) and \( x = 4 \), so its equation can be written as:  
$$y = a(x - 2)(x - 4)$$  
Substitute the point \( (3, 1) \) to find \( a \):  
$$1 = a(3 - 2)(3 - 4)$$  
$$1 = a(1)(-1)$$  
$$a = -1$$  
Thus, the equation is:  
$$y = -1(x - 2)(x - 4)$$  
Simplify:  
$$y = -x^2 + 6x - 8$$

### **Problem 3: Center of an Ellipse**
**Given:**  
Ellipse equation:  
$$x^2 + 4y^2 - 4x - 16y + 16 = 0$$

**Find the Center:**  
Rewrite the equation in standard form by completing the square:  
1. Group \( x \) and \( y \) terms:  
$$x^2 - 4x + 4y^2 - 16y = -16$$  
2. Complete the square for \( x \):  
$$x^2 - 4x = (x - 2)^2 - 4$$  
3. Complete the square for \( y \):  
$$4y^2 - 16y = 4(y^2 - 4y) = 4((y - 2)^2 - 4) = 4(y - 2)^2 - 16$$  
4. Substitute back:  
$$(x - 2)^2 - 4 + 4(y - 2)^2 - 16 = -16$$  
Simplify:  
$$(x - 2)^2 + 4(y - 2)^2 = 4$$  
Divide by 4:  
$$\frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1$$  
The center is at \( (2, 2) \).

### **Problem 4: Slope of a Tangent Line**
**Given:**  
- Line equation: \( y = mx - 5 \)  
- Circle equation: \( x^2 + y^2 = 1 \)  

**Find the Slope \( m > 0 \):**  
For the line to be tangent to the circle, the distance from the center of the circle to the line must equal the radius.  
The distance \( d \) from the origin \( (0, 0) \) to the line \( y = mx - 5 \) is:  
$$d = \frac{|m \cdot 0 - 1 \cdot 0 - 5|}{\sqrt{m^2 + 1}} = \frac{5}{\sqrt{m^2 + 1}}$$  
Set this equal to the radius \( r = 1 \):  
$$\frac{5}{\sqrt{m^2 + 1}} = 1$$  
Solve for \( m \):  
$$\sqrt{m^2 + 1} = 5$$  
$$m^2 + 1 = 25$$  
$$m^2 = 24$$  
$$m = \sqrt{24} = 2\sqrt{6}$$

### **Problem 5: Intersection Points of Hyperbola and Ellipse**
**Given:**  
- Hyperbola: \( x^2 - y^2 = 1 \)  
- Ellipse: \( x^2 + 4y^2 = 6 \)  

**Find Intersection Points:**  
Solve the system of equations:  
1. From the hyperbola: \( x^2 = y^2 + 1 \)  
2. Substitute into the ellipse:  
$$(y^2 + 1) + 4y^2 = 6$$  
$$5y^2 + 1 = 6$$  
$$5y^2 = 5$$  
$$y^2 = 1$$  
$$y = \pm 1$$  
3. Substitute \( y \) back into \( x^2 = y^2 + 1 \):  
For \( y = 1 \): \( x^2 = 2 \), so \( x = \pm \sqrt{2} \)  
For \( y = -1 \): \( x^2 = 2 \), so \( x = \pm \sqrt{2} \)  
Intersection points:  
$$(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)$$

### **Problem 6: Distance Between Branches of a Hyperbola**
**Given:**  
Hyperbola: \( x^2 - y^2 = 1 \)  

**Find the Distance Between Branches:**  
The hyperbola \( x^2 - y^2 = 1 \) has its branches along the \( x \)-axis. The distance between the vertices of the hyperbola is \( 2a \), where \( a \) is the distance from the center to a vertex.  
From the equation, \( a^2 = 1 \), so \( a = 1 \).  
Thus, the distance between branches is \( 2a = 2 \).
