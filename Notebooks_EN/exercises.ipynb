```python
#3
import numpy as np

def laplace_expansion(matrix):
  """Calculates the determinant of a 3x3 matrix using Laplace's expansion."""
  if matrix.shape != (3, 3):
    raise ValueError("Input must be a 3x3 matrix.")

  det = (
      matrix[0, 0] * (matrix[1, 1] * matrix[2, 2] - matrix[1, 2] * matrix[2, 1])
      - matrix[0, 1] * (matrix[1, 0] * matrix[2, 2] - matrix[1, 2] * matrix[2, 0])
      + matrix[0, 2] * (matrix[1, 0] * matrix[2, 1] - matrix[1, 1] * matrix[2, 0])
  )
  return det

# Example usage:
A = np.array([[2, 3, 1], [1, 4, 0], [3, 2, 1]])
det_A = laplace_expansion(A)
print("Determinant of A using Laplace's expansion:", det_A)
```

```python
#4
import numpy as np

def gauss_determinant(matrix):
  """Calculates the determinant of a 2x2 matrix using the Gauss method."""
  if matrix.shape != (2, 2):
    raise ValueError("Input must be a 2x2 matrix.")

  # Perform row operations (example: subtract a multiple of row 1 from row 2)
  factor = matrix[1, 0] / matrix[0, 0]
  matrix[1, :] = matrix[1, :] - factor * matrix[0, :]

  # Determinant is the product of diagonal elements
  det = matrix[0, 0] * matrix[1, 1]
  return det

# Example usage:
A = np.array([[12, 3], [-18, -4]])
det_A = gauss_determinant(A)
print("Determinant of A using Gauss method:", det_A)
```

```python
#7
def solve_linear_equations(eq1, eq2):
  """Solves a system of two linear equations using substitution.

  Args:
      eq1: A tuple representing the first equation in the form (a, b, c) for ax + by = c.
      eq2: A tuple representing the second equation in the form (a, b, c) for ax + by = c.

  Returns:
      A tuple containing the solutions for x and y, or None if there is no unique solution.
  """

  a1, b1, c1 = eq1  # Coefficients of equation 1
  a2, b2, c2 = eq2  # Coefficients of equation 2

  # Check for special cases (inconsistent or dependent systems)
  if a1 == 0 and b1 == 0:
    if c1 != 0:  # Inconsistent system
      return None
    else:  # Dependent system (infinitely many solutions)
      return None
  if a2 == 0 and b2 == 0:
    if c2 != 0:  # Inconsistent system
      return None
    else:  # Dependent system (infinitely many solutions)
      return None

  # Check if equations are equivalent (infinite solutions)
  if a1 * b2 == a2 * b1 and a1 * c2 == a2 * c1 and b1 * c2 == b2 * c1:
    return None  # Infinite solutions

  # Solve for x and y using substitution (avoiding premature use of y)
  if a1 != 0:  # Solve eq1 for x
    # x = (c1 - b1 * y) / a1  # We won't use this directly to avoid 'y'
    # Substitute into eq2: a2 * x + b2 * y = c2
    # a2 * ((c1 - b1 * y) / a1) + b2 * y = c2
    # Solve for y:
    y = (a1 * c2 - a2 * c1) / (a1 * b2 - a2 * b1)
    # Substitute y back into eq1 to find x:
    x = (c1 - b1 * y) / a1
  elif b1 != 0:  # Solve eq1 for y (if a1 is 0)
    y = (c1 - a1 * x) / b1  # substitute it into the other equation
    x = (c2 - b2*y) / a2
    y = (c1 - a1 * x) / b1

  else:  # if a1 and b1 are both 0 and c1 is 0
      return None

  return x, y

# Example usage:
eq1 = (3, -2, 5)  # 3x - 2y = 5
eq2 = (2, 3, 7)  # 2x + 3y = 7
solution = solve_linear_equations(eq1, eq2)

if solution:
  x, y = solution
  print("Solution: x =", x, ", y =", y)
else:
  print("No unique solution exists for this system of equations.")
```

```python
#8
import numpy as np

def cramer_rule(A, b):
  """Solves a system of linear equations using Cramer's rule.

  Args:
    A: The coefficient matrix (NumPy array).
    b: The constant vector (NumPy array).

  Returns:
    A NumPy array containing the solutions, or None if no unique solution exists.
  """

  n = A.shape[0]  # Number of variables
  det_A = np.linalg.det(A)

  if det_A == 0:
    return None  # No unique solution (determinant is 0)

  solutions = []
  for i in range(n):
    # Create a matrix by replacing the i-th column of A with b
    Ai = A.copy()
    Ai[:, i] = b

    # Calculate the solution for the i-th variable
    xi = np.linalg.det(Ai) / det_A
    solutions.append(xi)

  return np.array(solutions)

# Example usage for a 2x2 system:
A = np.array([[2, -3], [3, 5]])
b = np.array([7, 2])
solution = cramer_rule(A, b)

if solution is not None:
  print("Solution for 2x2 system:", solution)
else:
  print("No unique solution for 2x2 system.")

# Example usage for a 3x3 system:
A = np.array([[2, 1, -1], [1, -1, 2], [3, 0, -2]])
b = np.array([1, 4, -1])
solution = cramer_rule(A, b)

if solution is not None:
  print("Solution for 3x3 system:", solution)
else:
  print("No unique solution for 3x3 system.")
```

```python
#9
import numpy as np

def gaussian_elimination(A, b):
  """Solves a system of linear equations using Gaussian elimination.

  Args:
    A: The coefficient matrix (NumPy array).
    b: The constant vector (NumPy array).

  Returns:
    A NumPy array containing the solutions, or None if no unique solution exists.
  """

  n = A.shape[0]  # Number of variables
  augmented_matrix = np.concatenate([A, b.reshape(n, 1)], axis=1)  # Augment A with b

  # Forward elimination (transform to upper triangular form)
  for i in range(n):
    pivot = augmented_matrix[i, i]  # Choose pivot element
    if pivot == 0:
      # If pivot is 0, try to find a row below with a non-zero element in the same column
      for j in range(i + 1, n):
        if augmented_matrix[j, i] != 0:
          augmented_matrix[[i, j]] = augmented_matrix[[j, i]]  # Swap rows
          pivot = augmented_matrix[i, i]
          break
      else:
        return None  # No unique solution (pivot is 0 and no suitable row found)

    # Eliminate elements below the pivot
    for j in range(i + 1, n):
      factor = augmented_matrix[j, i] / pivot
      augmented_matrix[j, :] = augmented_matrix[j, :] - factor * augmented_matrix[i, :]

  # Back substitution (solve for variables)
  solutions = np.zeros(n)
  for i in range(n - 1, -1, -1):
    solutions[i] = (augmented_matrix[i, -1] - np.dot(augmented_matrix[i, i + 1:n], solutions[i + 1:])) / augmented_matrix[i, i]

  return solutions

# Example usage:
A = np.array([[1, 2, -2], [2, 1, 1], [3, 2, 1]])
b = np.array([4, 0, 1])
solution = gaussian_elimination(A, b)

if solution is not None:
  print("Solution using Gaussian elimination:", solution)
else:
  print("No unique solution found.")
```

```python
#10
import numpy as np

def solve_by_matrix_inversion(A, b):
  """Solves a system of linear equations using matrix inversion.

  Args:
    A: The coefficient matrix (NumPy array).
    b: The constant vector (NumPy array).

  Returns:
    A NumPy array containing the solutions, or None if the matrix is singular.
  """

  try:
    A_inv = np.linalg.inv(A)  # Calculate the inverse of A
    solution = np.dot(A_inv, b)  # Multiply the inverse by b to get the solution
    return solution
  except np.linalg.LinAlgError:
    return None  # Matrix is singular (no inverse exists)

# Example usage:
A = np.array([[1, 2, 3], [0, 2, 3], [0, 0, 3]])
b = np.array([5, 4, 3])
solution = solve_by_matrix_inversion(A, b)

if solution is not None:
  print("Solution using matrix inversion:", solution)
else:
  print("No unique solution found (matrix is singular).")
```
