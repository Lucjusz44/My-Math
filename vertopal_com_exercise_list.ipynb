## Tasks related to equations of lines on a plane

**Visualize each task.** - Python, GeoGebra, Wolfram Alpha, â€¦

-   The line passes through points $A(1, 2)$ and $B(3, 4)$. Find the
    equation of the line.
-   The line passes through point $A(1, 2)$ and is parallel to the line
    $y = 2x + 3$. Find the equation of the line.
-   The line passes through point $A(1, 2)$ and is perpendicular to the
    line $y = 2x + 3$. Find the equation of the line.
-   We have two lines $y = 2x + 3$ and $y = 3x + 2$. Find the
    intersection point of these lines and calculate the angle between
    them.
-   Write the equation of the line passing through point $A(1, 2)$ and
    parallel to the vector $\vec{v} = [2, 3]$.
-   We have the line $y = 2x + 3$. Find an example of a line
    perpendicular and parallel to it.
-   We have the line $y = 2x + 3$ and point $A(1, 2)$. Find the distance
    from point $A$ to the line.
-   The line intersects the coordinate axes at points $A(2, 0)$ and
    $B(0, 3)$. Find the equation of the line.
-   Calculate the angle between the line $y = x + 3$ and the $Ox$ axis.
-   Provide a vector perpendicular to the line $x + y + 1 = 0$.

```python
#1
import numpy as np
import matplotlib.pyplot as plt

# Points A and B
x1, y1 = 1, 2
x2, y2 = 3, 4

# Calculate slope (m)
m = (y2 - y1) / (x2 - x1)

# Calculate y-intercept (c)
c = y1 - m * x1

# Equation of the line: y = mx + c
print(f"Equation of the line: y = {m}x + {c}")

# Plotting the line
x = np.linspace(-5, 5, 100)  # Create x-values for the line
y = m * x + c  # Calculate corresponding y-values

plt.plot(x, y, label=f"y = {m}x + {c}")
plt.scatter([x1, x2], [y1, y2], color='red', label="Points A and B")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line through A(1, 2) and B(3, 4)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

# Point A
x1, y1 = 1, 2

# Slope (same as the given line)
m = 2

# Equation of the line: y - y1 = m(x - x1)
# y - 2 = 2(x - 1)
# y = 2x

print("Equation of the line: y = 2x")

# Plotting
x = np.linspace(-5, 5, 100)
y = m * x  # y = 2x

plt.plot(x, y, label="y = 2x")
plt.scatter(x1, y1, color='red', label="Point A")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line through A(1, 2) parallel to y = 2x + 3")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt

# Point A
x1, y1 = 1, 2

# Slope (negative reciprocal of the given line's slope)
m = -1/2

# Equation of the line: y - y1 = m(x - x1)
# y - 2 = -1/2(x - 1)
# y = -1/2x + 5/2

print("Equation of the line: y = -1/2x + 5/2")

# Plotting
x = np.linspace(-5, 5, 100)
y = m * x + (y1 - m * x1)  # y = -1/2x + 5/2

plt.plot(x, y, label="y = -1/2x + 5/2")
plt.scatter(x1, y1, color='red', label="Point A")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line through A(1, 2) perpendicular to y = 2x + 3")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt
import math

# Equations of the lines
eq1 = lambda x: 2*x + 3
eq2 = lambda x: 3*x + 2

# Find intersection point
# 2x + 3 = 3x + 2
# x = 1
# y = 2(1) + 3 = 5
intersection_x = 1
intersection_y = eq1(intersection_x)

print(f"Intersection point: ({intersection_x}, {intersection_y})")

# Calculate angle
m1 = 2  # Slope of the first line
m2 = 3  # Slope of the second line
angle_rad = math.atan(abs((m2 - m1) / (1 + m1 * m2)))
angle_deg = math.degrees(angle_rad)

print(f"Angle between the lines: {angle_deg:.2f} degrees")

# Plotting
x = np.linspace(-5, 5, 100)
y1 = eq1(x)
y2 = eq2(x)

plt.plot(x, y1, label="y = 2x + 3")
plt.plot(x, y2, label="y = 3x + 2")
plt.scatter(intersection_x, intersection_y, color='red', label="Intersection")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Intersection and Angle between Two Lines")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5
import numpy as np
import matplotlib.pyplot as plt

# Point A
x1, y1 = 1, 2

# Slope (from the vector)
m = 3/2

# Equation of the line: y - y1 = m(x - x1)
# y - 2 = 3/2(x - 1)
# y = 3/2x + 1/2

print("Equation of the line: y = 3/2x + 1/2")

# Plotting
x = np.linspace(-5, 5, 100)
y = m * x + (y1 - m * x1)  # y = 3/2x + 1/2

plt.plot(x, y, label="y = 3/2x + 1/2")
plt.scatter(x1, y1, color='red', label="Point A")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line through A(1, 2) parallel to vector [2, 3]")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6
import math

# Point A
x0, y0 = 1, 2

# Line equation: 2x - y + 3 = 0  (rewritten in the form Ax + By + C = 0)
A, B, C = 2, -1, 3

# Calculate distance
distance = abs(A * x0 + B * y0 + C) / math.sqrt(A**2 + B**2)

print(f"Distance from A(1, 2) to the line: {distance:.2f}")
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt

# Intercepts
a = 2  # x-intercept
b = 3  # y-intercept

# Equation of the line: x/a + y/b = 1
# x/2 + y/3 = 1
# 3x + 2y = 6
# y = -3/2x + 3

print("Equation of the line: y = -3/2x + 3")

# Plotting
x = np.linspace(-5, 5, 100)
y = -3/2 * x + 3

plt.plot(x, y, label="y = -3/2x + 3")
plt.scatter([a, 0], [0, b], color='red', label="Intercepts A and B")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line intersecting axes at A(2, 0) and B(0, 3)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#10
import numpy as np
import matplotlib.pyplot as plt

# Define the line equation: x + y + 1 = 0
def line_equation(x):
  return -x - 1  # Solve for y

# Define the perpendicular vector: [1, 1]
perp_vector = np.array([1, 1])

# Generate x values for plotting the line
x_values = np.linspace(-5, 5, 100)

# Calculate corresponding y values using the line equation
y_values = line_equation(x_values)

# Plot the line
plt.plot(x_values, y_values, label="x + y + 1 = 0")

# Plot the perpendicular vector starting at a point on the line
point_on_line = np.array([2, line_equation(2)]) # Choose x=2
plt.quiver(*point_on_line, *perp_vector, angles='xy', scale_units='xy', scale=1, color='red', label="Perpendicular Vector [1, 1]")

# Customize the plot
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line and Perpendicular Vector")
plt.grid(True)
plt.legend()
plt.show()
```
