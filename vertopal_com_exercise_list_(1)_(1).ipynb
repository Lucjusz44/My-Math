## Tasks related to conic sections

**Visualize each task.** - Python, GeoGebra, Wolfram Alpha, …

-   Find the equation of a circle with center at point $A(1,2)$ and
    radius $r=3$.
-   Find the equation of a parabola intersecting the $Ox$ axis at points
    $x=2$, $x=4$, and passing through point $y(3)=1$.
-   Find the center of the ellipse with the equation
    $x^2 + 4y^2 - 4x - 16y + 16 = 0$.
-   Find the slope ($m>0$) of the line $y=mx-5$ that is tangent to the
    circle with the equation $x^2 + y^2=1$.
-   Find the intersection points of the hyperbola $x^2 - y^2 = 1$ with
    the ellipse’s line $x^2 + 4y^2 = 6$.
-   For the given hyperbola $x^2 - y^2 = 1$, find the distance between
    its branches.

```python
#1
import matplotlib.pyplot as plt
import numpy as np

# Create data for the circle
theta = np.linspace(0, 2 * np.pi, 100)
x = 1 + 3 * np.cos(theta)
y = 2 + 3 * np.sin(theta)

# Plot the circle
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.title('Circle with Center (1, 2) and Radius 3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.xlim([-2, 4])
plt.ylim([-1, 5])
plt.gca().set_aspect('equal', adjustable='box')  # Ensure circle appears circular
plt.show()
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(1, 5, 100)

# Calculate corresponding y values using the equation
y = -(x - 2) * (x - 4)

# Plot the parabola
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.title('Parabola Intersecting x-axis at x=2 and x=4, Passing Through (3, 1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.scatter([2, 4, 3], [0, 0, 1], color='red', label='Key Points')  # Highlight key points
plt.legend()
plt.show()
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt

# Generate points for the ellipse
t = np.linspace(0, 2 * np.pi, 100)
x = 2 + 2 * np.cos(t)
y = 2 + 1 * np.sin(t)

# Plot the ellipse
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.title('Ellipse with Center (2, 2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.scatter(2, 2, color='red', label='Center (2, 2)')  # Highlight the center
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt

# Circle
theta = np.linspace(0, 2 * np.pi, 100)
x_circle = np.cos(theta)
y_circle = np.sin(theta)

# Tangent line
m = 2 * np.sqrt(6)
x_line = np.linspace(-2, 2, 100)
y_line = m * x_line - 5

# Plot
plt.figure(figsize=(5, 5))
plt.plot(x_circle, y_circle, label='Circle: x^2 + y^2 = 1')
plt.plot(x_line, y_line, label=f'Tangent Line: y = {m:.2f}x - 5')
plt.title('Circle and Tangent Line')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.xlim([-2, 2])
plt.ylim([-6, 2])
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hyperbola
x_hyperbola = np.linspace(-3, 3, 100)
y_hyperbola_pos = np.sqrt(x_hyperbola**2 - 1)
y_hyperbola_neg = -np.sqrt(x_hyperbola**2 - 1)

# Ellipse's line (rewritten as y = +/- sqrt((6 - x^2)/4))
x_ellipse = np.linspace(-np.sqrt(6), np.sqrt(6), 100)
y_ellipse_pos = np.sqrt((6 - x_ellipse**2) / 4)
y_ellipse_neg = -np.sqrt((6 - x_ellipse**2) / 4)

# Plot
plt.figure(figsize=(5, 5))
plt.plot(x_hyperbola, y_hyperbola_pos, label='Hyperbola: x^2 - y^2 = 1', color='blue')
plt.plot(x_hyperbola, y_hyperbola_neg, color='blue')
plt.plot(x_ellipse, y_ellipse_pos, label="Ellipse's Line: x^2 + 4y^2 = 6", color='red')
plt.plot(x_ellipse, y_ellipse_neg, color='red')
plt.title('Intersection of Hyperbola and Ellipse\'s Line')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()

# Highlight intersection points
intersection_points = [(np.sqrt(2), 1), (-np.sqrt(2), 1), (np.sqrt(2), -1), (-np.sqrt(2), -1)]
for point in intersection_points:
    plt.scatter(point[0], point[1], color='green', s=50)

plt.show()
```
